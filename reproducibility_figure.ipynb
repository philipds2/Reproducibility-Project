```r
library(ggplot2)
library(readxl)
```

```r
file <- download.file("https://doi.org/10.1371/journal.pbio.2004188.s001", destfile = "raw.xlsx", mode = "wb")
data <- read_xlsx("raw.xlsx", sheet = 2)
```

```r
ggplot(data = data, aes(x=`Y Coord`, y=`Theta PLV`)) +
  geom_point() +
  geom_smooth(color="black", method = "lm", se = FALSE) + 
  scale_x_continuous(breaks = seq(-100, 80, 20)) +
  scale_y_continuous(breaks = seq(-1.0, 0.4, 0.2)) +
  xlab("Posterior                          Y-coordinate                         Anterior") +
  theme(plot.title = element_text(hjust = 0.5)) +
  ylab("Effect of Age on PLV\n (Beta weight)") +
  ggtitle("Theta Age Effect Gradient") +
  theme(plot.title = element_text(hjust = 0.5)) +
  theme(panel.grid.major = element_blank(), panel.grid.minor = element_blank(),
        panel.background = element_blank(), axis.line = element_line(colour = "black"))
```
